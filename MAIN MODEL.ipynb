```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
```

```python
path=r'C:\Users\india\Downloads\gm_1\Dataset'
```

```python
tf.random.set_seed(42)
np.random.seed(42)
```

```python
model = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
datagen = keras.preprocessing.image.ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

train_generator = datagen.flow_from_directory(
    path,
    target_size=(128, 128),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

val_generator = datagen.flow_from_directory(
    path,
    target_size=(128, 128),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

```python
print(f"Training samples: {train_generator.samples}")
print(f"Validation samples: {val_generator.samples}")

from PIL import ImageFile

ImageFile.LOAD_TRUNCATED_IMAGES = True 

history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // 32,
    epochs=10,
    validation_data=val_generator,
    validation_steps=val_generator.samples // 32
)

test_loss, test_accuracy = model.evaluate(val_generator, steps=val_generator.samples // 32)
print(f"Test accuracy: {test_accuracy:.4f}")

model.save('simple_deepfake_model.h5')
print("Model saved as 'simple_deepfake_model.h5'")
```
